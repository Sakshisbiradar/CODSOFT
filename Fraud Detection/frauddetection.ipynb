<h1>Undersampling</h1>

<h3>Build a sample dataset with similar distributions</h3>

<h3>No of fraudulent transactions --> 492</h3>

```python
from imblearn.under_sampling import NearMiss
```

```python
import random
```

```python
legit_sample=legit.sample(n=492)
```

```python
newdataset=pd.concat([legit_sample, fraud] ,axis=0)
```

```python
newdataset['Class'].value_counts()
```

<h3>Now dataset has equal no. of values </h3>

```python
newdataset.groupby('Class').mean()
```

<h3>Split the data into train and test one containing targets and other features</h3>

```python
X=newdataset.drop(columns='Class', axis=1)
Y=newdataset['Class']
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score as score
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

</h3>Check accuracy</h3>

```python
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = score(Y_train, X_train_prediction)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
print('Accuracy score of train data : ', training_data_accuracy)
```

<h2>The accuracy came out to be 0.89 for test data and 0.92 for training data therefore our model is quite accurate as per the data we received.</h2>
